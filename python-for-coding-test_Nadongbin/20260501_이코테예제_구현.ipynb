```python
#문제1. 싱히좌우
```

```python
#1. 첫째줄에 공간의 크기를 나타내느 N이 주어진다.(1<=N<=100)
#2. 둘째 줄에 여행가A가 이동할 계획서 내용이 주어진다.(1<=이동횟수<=100)
#3. 첫째줄에 여행가 A가 최종적으로 도착할 지점의 좌표 (X,Y)를 공백으로 구분하여 출력한다.
```

```python
n= int(input())
x,y= 1,1
plans= input().split()
```

```python
print(plans)
print(type(plans))
```

```python
dx=[0, 0, -1, 1]
dy=[-1, 1, 0, 0]
#dx,dy[0]부터 L R U D조합

move_types=['L', 'R', 'U', 'D']
```

```python
for plan in plans: #plans에서 한글자씩 검사
    for i in range(len(move_types)): # L R U D중 뭔지 검사
        if plan == move_types[i]:
            nx = x + dx[i]
            ny = y + dy[i]
    if nx<1 or ny<1 or nx>n or ny > n:
        continue

    x, y = nx, ny
```

```python
print(x, y)
```
