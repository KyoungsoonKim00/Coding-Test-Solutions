```python
#재귀함수
def recursive_fuction(i):
    if i==100:
        return
        
    print(f'{i}번째 재귀 함수를 호출')
    recursive_fuction(i+1)
```

```python
recursive_fuction(1)
```

```python
def factorial_iterative(n): #반복적 구현
    result=1
    for i in range(1, n+1): #0, 1팩토리얼은 1
        result*=i
    return result
```

```python
def factorical_recursive(n):
    if n<=1:
        return 1
    return n* factorical_recursive(n-1)

```

```python

print(factorial_iterative(5))
print(factorical_recursive(5))
```

```python
def gcd(a, b):
    if a%b ==0:
        return b
    else:
        return gcd(b, a%b)

print( gcd(192, 162))
```

```python
'''
1. 탐색 시작 노드를 스택에 삽입하고 방문처리
2. 스택의 최상단 노드에 방문하지 않은 인접한 노드가 
하나라도 있으면 그 노드를 스택에 넣고 방문 처리.
방문하지 않은 인접 노드가 없으면 스택에서 최상단 노드를 꺼냄.
3. 더 이상 2번의 과정을 수행할 수 없을 떄까지 반복합니

'''
```

```python
#그래프를 다 탐색하는 순
def dfs(graph, v, visited):
    visted[v] = True
    print(v, end= '')
    for i in graph[v]:
        if not visted [i]:
            dfs(graph, i ,visted)
```

```python
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False] * 9
dfs(graph,1, visited)
```

```python
'''
1. 특정한 지점의 주변 상하좌우를 살펴본 뒤에 
주변 지점 중에서 값이 0이면서 아직 방문하지 않은 지점이 있다면 해당 지점을 방문
2. 방문한 지점에서 다시 상화좌우를 살펴보기(1반복)
3. 모든 노드에 대하여 1~2번의 과정을 반복하며 방문하지 않은 지점의 수를 카운트..
x,y>=0 (i in n j in m)
'''
```

```python
def dfs_ice(x, y):
    if x<=-1 or x>=n or y<=1 or y>=m:
        return False
    if graph[x][y] == 0:
        graph[x][y] = 1
        dfs_ice(x- 1, y)
        dfs_ice(x+ 1, y)
        dfs_ice(x, y-1)
        dfs_ice(x, y+1)
        return True
    return False
```
